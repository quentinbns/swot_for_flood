```python
CONDITIONS_WORLDCOVER = {
    "urban": 50,
    "forest": 10,
    "permament_water": 80
}

def get_condition(WC_array):
    """
    Get the condition of the world cover array
    """
    condition_urban = WC_array['band_1'].values == CONDITIONS_WORLDCOVER["urban"]
    condition_forest = WC_array['band_1'].values == CONDITIONS_WORLDCOVER["forest"]
    condition_permament_water = WC_array['band_1'].values == CONDITIONS_WORLDCOVER["permament_water"]
    condition_open = np.logical_and(~ condition_forest,
        np.logical_and(
            ~ condition_urban,
            ~ condition_permament_water)
        )
    return condition_urban, condition_forest, condition_permament_water, condition_open
```

```python
Chinon_urban_global, Chinon_forest_global, Chinon_permament_water_global, Chinon_open_global = get_condition(Chinon_project.swot_collection.ESA_WC)
Chinon_urban_flood, Chinon_forest_flood, Chinon_permament_water_flood, Chinon_open_flood = get_condition(Chinon_project.swot_collection.ESA_WC_FLOOD)
Chinon_urban_control, Chinon_forest_control, Chinon_permament_water_control, Chinon_open_control = get_condition(Chinon_project.swot_collection.ESA_WC_CONTROL)

Ohio_urban_global, Ohio_forest_global, Ohio_permament_water_global, Ohio_open_global = get_condition(Ohio_project.swot_collection.ESA_WC)
Ohio_urban_flood, Ohio_forest_flood, Ohio_permament_water_flood, Ohio_open_flood = get_condition(Ohio_project.swot_collection.ESA_WC_FLOOD)
Ohio_urban_control, Ohio_forest_control, Ohio_permament_water_control, Ohio_open_control = get_condition(Ohio_project.swot_collection.ESA_WC_CONTROL)

EMSR692_urban_global, EMSR692_forest_global, EMSR692_permament_water_global, EMSR692_open_global = get_condition(EMSR692_project.swot_collection.ESA_WC)
EMSR692_urban_flood, EMSR692_forest_flood, EMSR692_permament_water_flood, EMSR692_open_flood = get_condition(EMSR692_project.swot_collection.ESA_WC_FLOOD)
EMSR692_urban_control, EMSR692_forest_control, EMSR692_permament_water_control, EMSR692_open_control = get_condition(EMSR692_project.swot_collection.ESA_WC_CONTROL)

PortoAlegre_urban_global, PortoAlegre_forest_global, PortoAlegre_permament_water_global, PortoAlegre_open_global = get_condition(PortoAlegre_project.swot_collection.ESA_WC)
PortoAlegre_urban_flood, PortoAlegre_forest_flood, PortoAlegre_permament_water_flood, PortoAlegre_open_flood = get_condition(PortoAlegre_project.swot_collection.ESA_WC_FLOOD)
PortoAlegre_urban_control, PortoAlegre_forest_control, PortoAlegre_permament_water_control, PortoAlegre_open_control = get_condition(PortoAlegre_project.swot_collection.ESA_WC_CONTROL)
```
